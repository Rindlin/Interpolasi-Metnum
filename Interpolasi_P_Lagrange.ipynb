```python
import numpy as np
import matplotlib.pyplot as plt

# Titik data yang baru
x_points = [5, 10, 15, 20, 25, 30, 35, 40]
y_points = [40, 30, 25, 40, 18, 20, 22, 15]

# Fungsi interpolasi Lagrange
def lagrange_interpolation(x_points, y_points, x):

    if len(x_points) != len(y_points):
        raise ValueError("Panjang x_points dan y_points harus sama.")

    n = len(x_points)
    result = 0.0

    for i in range(n):
        term = y_points[i]
        for j in range(n):
            if i != j:
                term *= (x - x_points[j]) / (x_points[i] - x_points[j])
        result += term

    return result

# Menghitung interpolasi untuk array x pada rentang 5 <= x <= 40
x_new = np.linspace(5, 40, 100)
y_new = [lagrange_interpolation(x_points, y_points, xi) for xi in x_new]

# Membuat plot
plt.figure(figsize=(10, 8))

# Plot titik data asli
plt.plot(x_points, y_points, 'ro', label='Titik data')

# Plot hasil interpolasi Lagrange
plt.plot(x_new, y_new, 'b-', label='Interpolasi Lagrange')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolasi Polinomial Lagrange')
plt.grid(True)
plt.show()
```
